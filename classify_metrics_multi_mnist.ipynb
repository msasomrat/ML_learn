```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('digit_train.csv')
```

```python
df
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier


from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score,confusion_matrix

from sklearn.metrics import precision_score,recall_score,f1_score

from sklearn.metrics import classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = LogisticRegression()
```

```python
model2 = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model2.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
prediction2 = model2.predict(x_test)
```

```python
accuracy_score(y_test, prediction)
```

```python
accuracy_score(y_test, prediction2)
```

```python
confusion_matrix(y_test, prediction)
```

```python
confusion_matrix(y_test, prediction2)
```

```python
precision_score(y_test, prediction, average=None)
```

```python
precision_score(y_test, prediction, average='weighted')
```

```python
precision_score(y_test, prediction2, average='weighted')
```

```python
recall_score(y_test, prediction, average='weighted')
```

```python
recall_score(y_test, prediction2, average='weighted')
```

```python
f1_score(y_test, prediction, average='weighted')
```

```python
f1_score(y_test, prediction2, average='weighted')
```

```python
report_dict = classification_report(y_test, prediction, output_dict=True)
```
