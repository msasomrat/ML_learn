```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
df
```

```python
df.drop(['petal length (cm)', 'petal width (cm)'], axis='columns' , inplace=True)
```

```python
df
```

```python
df
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters= 3)
k_means
```

```python
y_predicted = k_means.fit_predict(df)
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
k_means.cluster_centers_
```

```python
df1 = df[df.cluster ==0]
df2 = df[df.cluster ==1]
df3 = df[df.cluster ==2]
```

```python

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'green', label = 'sepal')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'red', label = 'sepal')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'], color = 'black', label = 'sepal')

plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], color = 'purple', marker = '*', label = 'centroid')

plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
scale.fit(df[['sepal length (cm)']])
df['sepal length (cm)'] = scale.transform(df[['sepal length (cm)']])
```

```python
scale.fit(df[['sepal width (cm)']])
df['sepal width (cm)'] = scale.transform(df[['sepal width (cm)']])
```

```python
df
```

```python
k_means = KMeans(n_clusters= 3)
y_predicted = k_means.fit_predict(df[['sepal length (cm)', 'sepal width (cm)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
k_means.cluster_centers_
```

```python
df1 = df[df.cluster ==0]
df2 = df[df.cluster ==1]
df3 = df[df.cluster ==2]
```

```python

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'green', label = 'sepal')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color = 'red', label = 'sepal')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'], color = 'black', label = 'sepal')

plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], color = 'purple', marker = '*', label = 'centroid')

plt.legend()
```

```python
k_range = range(1,10)
sse = []

for k in k_range:
    k_means = KMeans(n_clusters= k)
    k_means.fit(df[['sepal length (cm)', 'sepal width (cm)']])
    sse.append(k_means.inertia_)
```

```python
sse
```

```python
plt.plot(k_range, sse)
```
