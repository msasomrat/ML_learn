```python
b = -50
m = 27.82809103
lr = 0.001
all_b = []
all_cost = []
```

```python
epochs = 100
```

```python
for i in range(epochs):
    slope = 0
    cost = 0
    for j in range(x.shape[0]):
        slope = slope - 2*(y[j] - (m * x[j]) -  b)
        cost = cost + (y[j] - m * x[j] -b) ** 2

    b = b - (lr * slope)
    all_b.append(b)
    all_cost.append(cost)
    y_pred = m * x + b
    plt.plot(x,y_pred)
```
