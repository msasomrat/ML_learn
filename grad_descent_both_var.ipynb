```python

b = -520
m = 600
lr = 0.001
all_b = []
all_m = []
all_cost = []

epochs = 30

for i in range(epochs):
    slope_b = 0
    slope_m = 0
    cost = 0
    for j in range(X.shape[0]):
        slope_b = slope_b - 2*(y[j] - (m * [j]) -  b)
        slope_m = slope_m - 2*(y[j] - (m * X[j]) -  b)*X[j]
        cost = cost + (y[j] - m * X[j] -b) ** 2

    b = b - (lr * slope_b)
    m = m - (lr * slope_m)
    all_b.append(b)
    all_m.append(m)
    all_cost.append(cost)
```

```python
fig, ax = plt.subplots(figsize=(9,5))
#fig.set_tight_layout(True)

x_i = np.arange(-3, 3, 0.1)
y_i = x_i*(-27) -150
ax.scatter(X, y)
line, = ax.plot(x_i, x_i*50 - 4, 'r-', linewidth=2)

def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i*all_m[i] + all_b[i])
    ax.set_xlabel(label)
    # return line, ax

anim = FuncAnimation(fig, update, repeat=True, frames=epochs, interval=500)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation

num_epochs = list(range(0, 30))
all_b = [i * 5 for i in num_epochs]  # Example data

fig = plt.figure(figsize=(20, 20))
axis = plt.axes(xlim=(0, 30), ylim=(-10, 160))

line, = axis.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    label = 'epoch {0}'.format(i + 1)
    xdata = num_epochs[:i + 1]
    ydata = all_b[:i + 1]
    line.set_data(xdata, ydata)
    axis.set_xlabel(label)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(num_epochs), repeat=False, interval=500)

f = "animation6.gif"
writergif = animation.PillowWriter(fps=2)
anim.save(f, writer=writergif)
```
