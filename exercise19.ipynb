```python
df3.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace= True
)
```

```python
df4 = pd.get_dummies(df3, drop_first=True)
```

```python
df4
```

```python
x = df4.drop('HeartDisease', axis = 'columns')
```

```python
x
```

```python
y = df4.HeartDisease
```

```python
y
```
