```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df
```

```python
df.shape
```

```python
df.drop([['Address', 'Date', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude']], inplace= True)
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
```

```python
df = df[cols_to_use]
```

```python
df
```

```python
df.nunique()
```

```python
df.isna().any()
```

```python
df.isna().sum()
```

```python
df[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']] = df[['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']].fillna(0)
```

```python
df
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.dropna(inplace= True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
x = df.drop('Price', axis = 'columns')
y = df['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression()
```

```python
model2.fit(x_train, y_train)
```

```python
model2.score(x_train, y_train)
```

```python
from sklearn.svm import SVC
model3 = SVC()
```

```python
model3.fit(x_train, y_train)
```

```python
model3.score(x_test, y_test)
```

```python
from sklearn.linear_model import Lasso
```

```python
model4 = Lasso(alpha= 50, max_iter= 1000, tol=0.1)
```

```python
model4.fit(x_train, y_train)
```

```python
model4.score(x_test, y_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model5 = Ridge(alpha= 50, max_iter= 10000, tol= 0.1)
```

```python
model5.fit(x_train, y_train)
```

```python
model5.score(x_test, y_test)
```

```python
model5.score(x_train, y_train)
```
