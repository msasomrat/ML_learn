```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
covariance_matrix = np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
```

```python
covariance_matrix
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from mpl_toolkits.mplot3d.art3d import Line3D


fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')



ax.scatter(df['feature1'], df['feature2'], df['feature3'], color='blue', alpha=0.2, s=80)


mean_point = [np.mean(df['feature1']), np.mean(df['feature2']), np.mean(df['feature3'])]
ax.scatter(*mean_point, color='red', alpha=0.5, s=100)


for v in eigen_vectors.T:
    ax.plot([mean_point[0], v[0]], [mean_point[1], v[1]], [mean_point[2], v[2]], color='r', linestyle='--', linewidth=2)

ax.set_xlabel('x_values')
ax.set_ylabel('y_values')
ax.set_zlabel('z_values')

plt.title('Eigenvectors')
plt.show()
```

```python
df = pd.DataFrame({
    'feature1': np.random.rand(10),
    'feature2': np.random.rand(10),
    'feature3': np.random.rand(10),
    'target': np.random.randint(0, 2, 10)
})

# Principal components matrix (example)
pc = np.random.rand(3, 2)  # 3 features, 2 principal components

# Perform the matrix multiplication
transformed_df = np.dot(df.iloc[:, 0:3], pc)

# Create a new DataFrame with the transformed data
new_df = pd.DataFrame(transformed_df, columns=['PC1', 'PC2'])
new_df['target'] = df['target'].values

# Display the first few rows of the new DataFrame
print(new_df.head())
```

```python

new_df['target'] = new_df['target'].astype('str')
fig = px.scatter(x=new_df['PC1'],
                 y=new_df['PC2'],
                 color=new_df['target'],
                 color_discrete_sequence=px.colors.qualitative.G10
                )

fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```
