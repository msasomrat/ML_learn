```python
merge_df['PURCHASE_DATE'] = pd.to_datetime(merge_df['PURCHASE_DATE'], format="%d/%m/%Y")

# Filter purchases made in the first 15 days of March 2019
march_purchases_df = merge_df[(merge_df['PURCHASE_DATE'] >= '01/03/2019') &
                               (merge_df['PURCHASE_DATE'] <= '15/03/2019')]
```
