```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
digit = load_digits()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = digit.data
x
```

```python
y = digit.target
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2)
```

```python
len(x_train)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
def Score(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    return model.score(x_test, y_test)
```

```python
Score(LogisticRegression(max_iter=2000), x_train, x_test, y_train, y_test)
```

```python
Score(SVC(), x_train, x_test, y_train, y_test)
```

```python
Score(RandomForestClassifier(n_estimators=40), x_train, x_test, y_train, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=3)
```

```python
for train_index, test_index in skf.split(digit.data, digit.target):
    print(train_index, test_index)
```

```python
score_lr = []
score_svm = []
score_rf = []

for train_index, test_index in skf.split(digit.data, digit.target):
    x_train, x_test, y_train, y_test = digit.data[train_index], digit.data[test_index], \
    digit.target[train_index], digit.target[test_index]

    print(Score(LogisticRegression(max_iter=2000), x_train, x_test, y_train, y_test))
    print(Score(SVC(), x_train, x_test, y_train, y_test))
    print(Score(RandomForestClassifier(n_estimators=40), x_train, x_test, y_train, y_test))
```

```python
score_lr = []
score_svm = []
score_rf = []

for train_index, test_index in skf.split(digit.data, digit.target):
    x_train, x_test, y_train, y_test = digit.data[train_index], digit.data[test_index], \
    digit.target[train_index], digit.target[test_index]

    score_lr.append(Score(LogisticRegression(max_iter=2000), x_train, x_test, y_train, y_test))
    score_svm.append(Score(SVC(), x_train, x_test, y_train, y_test))
    score_rf.append(Score(RandomForestClassifier(n_estimators=40), x_train, x_test, y_train, y_test))
```
