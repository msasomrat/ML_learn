```python


# Assuming titanic is your DataFrame and it's already loaded
sn.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Did not survive', shade=True)
sn.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label='Survived', shade=True)

plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age by Survival Status')
plt.legend()
plt.show()
```

```python
sn.displot(titanic[titanic['Survived'] == 0], x='Age', kind='kde', fill=True, label='Did not survive')
sn.displot(titanic[titanic['Survived'] == 1], x='Age', kind='kde', fill=True, label='Survived')

plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution of Age by Survival Status')
plt.legend()
plt.show()
```
