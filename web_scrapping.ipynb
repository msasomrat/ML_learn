```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

# URL of the page to scrape
url = 'https://medex.com.bd/companies/2/aci-limited/brands'

# Make the request with headers and get the response
response = requests.get(url, headers=headers).text
```

```python
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
divs_with_urls = soup.find_all('div', {'class': 'your-class-name'})  # Adjust 'class' attribute as needed

for div in divs_with_urls:
    # Assuming URLs are in <a> tags within the <div>
    links = div.find_all('a', href=True)
    
    for link in links:
        drug_url = link['href']
        print(f"Found URL: {drug_url}")
        
        # Now fetch the URL if needed
        try:
            drug_response = requests.get(drug_url)
            drug_response.raise_for_status()
            drug_soup = BeautifulSoup(drug_response.text, 'html.parser')
            
            # Continue processing drug_soup as needed
            
        except requests.exceptions.RequestException as e:
            print(f"Error fetching URL {drug_url}: {e}")
```
