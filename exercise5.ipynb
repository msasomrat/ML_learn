```python
sub_df = df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years','salary']]
sub_df
```

```python
dummies = pd.get_dummies(sub_df.salary, prefix='salary')
df_dummies = pd.concat([sub_df, dummies], axis='columns')
df_dummies
```

```python
df_dummies.drop('salary', axis='columns', inplace= True)
```

```python
df_dummies
```

```python
x= df_dummies
```

```python
y = df.left
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
log_reg.predict(x_test)
```

```python
log_reg.score(x_train, y_train )
```

```python
log_reg.score(x_test, y_test )
```
