```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('titanic.csv', usecols = ['Age', 'Fare', 'Survived'])
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
df.corr()
```

```python
df.isna().sum()
```

```python
x = df.drop(columns='Survived')
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 2)
```

```python
x_train
```

```python
x_test
```

```python
model = SimpleImputer()
```

```python
x_train_trf = model.fit_transform(x_train)
x_test_trf = model.transform(x_test)
```

```python
x_test_trf
```

```python
x_test_trf.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2 = LogisticRegression()
```

```python
model2.fit(x_train_trf, y_train)
```

```python
prediction = model2.predict(x_test_trf)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prediction)
```

```python
mi = MissingIndicator()
```

```python
mi.fit(x_train)
```

```python
mi.features_
```

```python
mi.features
```

```python
x_train_missing = mi.transform(x_train)
x_train_missing
```

```python
x_test_missing = mi.transform(x_test)
```

```python
x_test_missing
```

```python
x_train['Age_NA'] = x_train_missing
x_train
```

```python
x_test['Age_NA'] = x_test_missing
x_test
```

```python
x_train_trf2 = model.fit_transform(x_train)
```

```python
x_test_trf2 = model.transform(x_test)
```

```python
model2.fit(x_train_trf2, y_train)
```

```python
prediction = model2.predict(x_test_trf2)
```

```python
accuracy_score(y_test, prediction)
```

```python
model3 = SimpleImputer(add_indicator=True)
```

```python
x_train = model3.fit_transform(x_train)
```

```python
x_test = model3.transform(x_test)
```

```python
prediction = model2.predict(x_test_trf2)
```
